```python
# menampilkan tabel 
from tabulate import tabulate

def show_data(table):
    tab = tabulate(tabular_data=table,
                   headers=table.keys(),
                   tablefmt="psql",
                   numalign="center")
    print(tab)
```

```python
# membuat dummy data table library 
def library(n_cust, is_print):
    tabel = {}
    tabel["library_id"] = [i+1 for i in range(n_cust)]
    tabel["library_name"] = [fake.company() for i in range(n_cust)]
    tabel["address"] = [fake.address() for i in range(n_cust)]
    if is_print:
        show_data(tabel)
        
    return tabel
```

```python
library = library(100, True)
```

```python
# meyimpan file kedalam csv 
import csv
```

```python
def save_to_csv(data, nama_file):
    '''
    Fungsi untuk menyimpan data dummy ke csv

    args:
        - data (list)     : list of dictionary data yang akan dijadikan csv
        - nama_file (str) : nama untuk file csv

	return:
		- None
    '''

    # Membuat file csv
    with open(file = f"{nama_file}.csv", mode = 'w', newline = '') as csv_file:
        # Membuat writer csv
        writer = csv.writer(csv_file)

        # write header csv
        writer.writerow(list(data.keys()))

        # mengetahui panjang data
        len_data = len(list(data.items())[0][1])

        # write data ke file csv
        for i in range(len_data):
            row = []
            for key in data.keys():
                row.append(data[key][i])
            writer.writerow(row)
```

```python
save_to_csv(data = library,
            nama_file = 'library')
```

```python
# membuat tabel category 
kategori_buku = [
    "Fiksi", "Non-Fiksi", "Biografi", "Fantasi", "Sains", "Teknologi",
    "Sejarah", "Romantis", "Misteri", "Thriller", "Horror", "Psikologi",
    "Agama", "Pendidikan", "Bisnis", "Anak-anak", "Sastra", "Self-Improvement"
]

def category(is_print):
    n_cust = len(kategori_buku)
    tabel = {}
    tabel["category_id"] = [i+1 for i in range(n_cust)]
    tabel["category"] = kategori_buku.copy()
    if is_print:
        show_data(tabel)
        
    return tabel
```

```python
#memampilkan judul buku 
def fake_judul():
    return ' '.join(fake.words(nb=random.randint(2, 5))).title()

# membuat tabel book 
def book(n_cust, is_print):
    tabel = {}
    tabel["book_id"] = [f" B - {i+1:03}" for i in range(n_cust)]
    tabel["category_id"] = [random.choice (category["category_id"])  for i in range(n_cust)]
    tabel["judul"] = [fake_judul() for i in range(n_cust)]
    tabel["penulis"] = [fake.name() for i in range(n_cust)]
    if is_print:
        show_data(tabel)
        
    return tabel
```

```python
# membuat tabel users 
def users(n_cust, is_print):
    tabel = {
        "users_id" : [ i+1 for i in range(n_cust)],
        "name" : [],
        "email" : [],
        "password" :[f"{fake.words(nb=1)[0]}{fake.random_int(max=99)}" for i in range(n_cust)] }
    for _ in range(n_cust):
        nama = fake.name()
        nama_1 = "".join(nama.split()[:2])
        nama_bersih = nama_1.lower().replace(" ", "").replace(".", "")
        email = f"{nama_bersih}@gmail.com"
    

        tabel["name"].append(nama)
        tabel["email"].append(email)
    
    if is_print:
        show_data(tabel)
    return tabel 
```

```python
# membuat tabel loan 
def loan(n_cust, is_print):
    tabel = {}
    tabel["loan_id"] = [i + 1 for i in range(n_cust)]
    tabel["users_id"] = [random.choice(users["users_id"]) for i in range(n_cust)]
    tabel["library_id"] = [random.choice(library["library_id"]) for i in range(n_cust)]
    tabel["book_id"] = [random.choice(book["book_id"]) for i in range(n_cust)]

    loan_dates = [fake.date_between(start_date=start_date, end_date=end_date) for _ in range(n_cust)]
    tabel["loan_date"] = loan_dates

    due_dates = [loan_date + timedelta(days=14) for loan_date in loan_dates]
    tabel["due_date"] = due_dates
    today = datetime.today().date()
    return_dates = []
    for loan_date, due_date in zip (loan_dates, due_dates):
        if random.random() < 0.7:
              # 70% kemungkinan punya return_date
            return_dates.append(fake.date_between(start_date=loan_date + timedelta(days=1),
                                                     end_date=loan_date + timedelta(days= 14)))
        elif due_date < today: 
            return_dates.append(due_date)
        else:
            return_dates.append(None)  # sisanya None
    tabel["return_date"] = return_dates

    status_loan_id = []
    for return_date, due_date in zip(return_dates, due_dates):
        if return_date is not None:
            status_loan_id.append(2) # returned
    
        elif due_date < today :
            status_loan_id.append(2)
        else:
            status_loan_id.append(1)  # active
        
    tabel["status_loan_id"] = status_loan_id

    if is_print:
        show_data(tabel)
    return tabel
```

```python
loan = loan (100, True)
```

```python
save_to_csv(data = loan,
            nama_file = 'loan')
```

```python
from datetime import datetime, timedelta

start_date = datetime.strptime("2025-01-01", "%Y-%m-%d").date()
end_date = datetime.strptime("2025-12-31", "%Y-%m-%d").date()
```

```python
def hold_1(n_cust, is_print):
    tabel = {
        "hold_id":[i+1 for i in range(n_cust)],
        "users_id" : [ random.choice(users["users_id"]) for i in range(n_cust)],
        "library_id": [random.choice(library["library_id"]) for i in range(n_cust)],
        "book_id" : [random.choice(book["book_id"]) for i in range(n_cust)], 
        "hold_date" : [],
        "available_date" :[],
        "expired_date": [],
        "pickup_date" : [],
        "status_hold_id" : []
    }
    today = datetime.today().date()
    
    for i in range(n_cust):
        hold_date = fake.date_between(start_date = start_date, end_date= end_date)
        available_date = fake.date_between(start_date = hold_date+timedelta(days=1), end_date= hold_date + timedelta(days=56))
        expired_date = available_date + timedelta(days = 7)
        #pickup date
        if random.random() < 0.7:  # 70% kemungkinan 
            pickup_date = fake.date_between(start_date=available_date + timedelta(days = 1),
                                                      end_date= available_date + timedelta(days=6))
        else:
            pickup_date = None  # sisanya None
        #mengecek status hold
        if pickup_date is not None:
            status_hold_id = 3# fullfield
        else:
            if  today > expired_date:
                status_hold_id= 2  # expierd
            else:
                status_hold_id = 1 # waiting
        tabel["pickup_date"].append(pickup_date)
        tabel["hold_date"].append(hold_date)
        tabel["expired_date"].append(expired_date)
        tabel["available_date"].append(available_date)
        tabel["status_hold_id"]. append(status_hold_id)
    if is_print: 
        show_data(tabel)
    return tabel
```

```python
hold_1= hold_1(100, True)
```

```python
save_to_csv(data = hold_1,
            nama_file = 'hold')
```
